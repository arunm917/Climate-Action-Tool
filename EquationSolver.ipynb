```python
equations_df = pd.read_excel('equations')
```

```python
equations_df
```

```python
coeff_matrix_df = equations_df.iloc[:,0:14]
coeff_matrix_df
```

```python
coeff_matrix = coeff_matrix_df.values.tolist()
```

```python
print(coeff_matrix)
```

```python
constants_df = equations_df['Constant']
constants_df
```

```python
constants = constants_df.values.tolist()
# constants = constants*-1
print(constants)
```

```python
x = np.linalg.solve(coeff_matrix, constants)
print(x)
```

```python
eigenvalues = np.linalg.eigvals(coeff_matrix)
print(eigenvalues)
```

```python
det = np.linalg.det(coeff_matrix)
print(det)
```

```python
# solve the system of linear equations
x, residuals, _, _ = np.linalg.lstsq(coeff_matrix, constants, rcond=None)
# print the solution
print(x)
print(residuals)
```
